```python
df = pd.read_csv('/kaggle/input/hiring/hiring.csv')
df
```

```python
def digit_str_to_int(digit_as_str):
    switcher = {
        "zero": 0,
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9,
        "ten": 10,
        "eleven": 11
    }

    # get() method of dictionary data type returns
    # value of passed argument if it is present
    # in dictionary otherwise the second argument will
    # be assigned as default value of passed argument
    digit_as_int = switcher.get(digit_as_str, -1)
    return digit_as_int

# assuming your digit strings are in 'digits' column,
# you can apply the transformation below
#df['digit_int'] = df['digits'].apply(digit_str_to_int)
```

Imputing to fill empty values, using median test score and setting NA to zero in experience and mapping from string to integer

```python

df.experience = df.experience.fillna('zero')
df["experience"] = df.experience.astype(str)
df.experience = df.experience.apply(digit_str_to_int)
import math
median_test_score = math.floor(df['test_score(out of 10)'].mean())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

Fitting a MLR model

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```
